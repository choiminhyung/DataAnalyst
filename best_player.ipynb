```python
columns_select = ['팀', '이름', '생일','일자', '상대','타수','안타','홈런', '루타', '타점','볼넷', '사구', '희비']
data = raw[columns_select]
data 
```

# 선수별 기록 집계하기

```python
data_player = pd.pivot_table(data, index = ['팀','이름','생일'], values = ['타수','안타','홈런','루타','타점','볼넷','사구','희비'], aggfunc= 'sum')
data_player
```

# 타수가 0이거나 너무 적은 데이터는 동일한 기준으로 비교가 어려움
판다스에 내장엔 histogrm으로 타수를 파악해보자.

```python
data_player['타수'].hist()

#0부터 50까지 사이에 150개 가까운 데이터가 있다. 많은 수의 타자가 0~50까지에 존재한다.
```

```python
#타수가 50보다 큰 사람만 지정해보자
```

```python
cond = data_player['타수'] > 50
data_player[cond]
```

```python
#멀티인덱스의 경우 특정 인덱스 조건으로 선택하기가 쉽지않다. 인덱스를 컬럼으로 
```

```python
data_player = data_player[cond].reset_index()
data_player
```

```python
# 타율/출루율/장타율/OPS를 계산하는 함수를 만들어보겠습니다. 
# 데이터 프레임을 입력하면, 해당 데이터 프레임에서 인덱스별 실적을 계산하여 반환해주는 함수입니다. 

# 타율 : 타격에 성공해서 진루하는 비율 --> 안타 / 타수
# 출루율: 살아서 진루하는 비율 -->  (안타+볼넷+몸에맞는볼)/(타수+볼넷+몸에맞는볼+희생플라이)
#  장타율 : 타율에 진루한 베이스 가중치 추가 -->   루타 / 타수


def cal_hit(df):
  
    df['타율'] = df['안타'] / df['타수']
    df['출루율'] = (df['안타'] + df['볼넷'] + df['사구']) / (df['타수'] + df['사구'] + df['희비'])
    df['장타율'] = df['루타'] / df['타수']
    df['OPS'] = df['출루율'] + df['장타율']
    return df
```

```python
# data_player에 있는 선수별 실적을 이용해 타율, 출루율, 장타율, OPS 를 계산한 데이터프레임을 가져오겠습니다. 

player_stat = cal_hit(data_player)  #타율, 출루율, 장타율, OPS 컬럼이 생김
player_stat.head()
```

```python
# 출루율/장타율/OPS/타율 기준으로 KBO 최고 타자는 누구인지 성적순으로 정렬해보겠습니다. 
# 출루율을 기준으로 정렬을 하며, 만약 동률일 경우 그다음 기준인 장타율을, 이후에는 OPS, 타율을 기준으로 정렬하였습니다. 
```

```python
player_stat =player_stat.sort_values(by = ['출루율','장타율','OPS','타율'], ascending= False)
player_stat
```

```python
#인덱스 번호가 의미가 없어서 reset_index로 기존인덱스를 컬럼으로 바꿔주는데, 인덱스가 의미없으므로 초기화해준다.
```

```python
player_stat = player_stat.reset_index(drop = True)
player_stat.head(20)
```
