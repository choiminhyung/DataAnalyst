```python
def find_dis_sk(page_no) :
    url = f'https://finance.naver.com/item/board.nhn?code=017670&page={page_no}'
    response = requests.get(url, headers =headers )
    response_bs = bs(response.text, 'lxml')
    table =response_bs.select('table')
    sk_table = pd.read_html(str(table), encoding = 'cp949')
    df_sk = sk_table[1]
    return df_sk


# url ='https://finance.naver.com/item/board.nhn?code=017670'
# response = requests.get(url, headers =headers )
```
